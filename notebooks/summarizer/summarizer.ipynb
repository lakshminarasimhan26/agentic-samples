```python
import gradio as gr
from openai import OpenAI
```

```python
# Initialize the OpenAI client to connect to the local Ollama instance
client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama', # The api_key is required by the library but unused by Ollama
)

system_message = """
You are a professional summarizer. 
You will be given a text and you will need to summarize it in a concise manner 
with no more than 100 words. You should always respond in English 
and if you do not have certain info, say so.
""" 
```

```python
def summarizer(message, history):
    # Define the messages for the chat
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": message}
    ]

    # Generate a chat completion using the 'llama3.2' model
    response = client.chat.completions.create(
        model="llama3.2", # Use the name of the model you pulled
        messages=messages,
        temperature=0.7 # Optional: adjust creativity
    )

    return response.choices[0].message.content
```

```python
s = gr.ChatInterface(fn=summarizer, title="Summarizer Bot")
s.launch()
```
